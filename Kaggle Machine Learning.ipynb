```python
#Filling the missing values of Age:
train_df['AgeFill'] = train_df['Age']

# Populate AgeFill
train_df['AgeFill'] = train_df['AgeFill'] \
                        .groupby([train_df['Sex'], train_df['Pclass']]) \
                        .apply(lambda x: x.fillna(x.median()))
```
